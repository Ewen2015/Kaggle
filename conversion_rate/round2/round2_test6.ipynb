```python
import pandas as pd

wd = '/Users/ewenwang/Documents/practice_data/conversion_rate/'
file = ['round2_keras_train_7.txt', 'round2_keras_test_a.txt']

train = pd.read_csv(wd+file[0], sep=' ')
test = pd.read_csv(wd+file[1], sep=' ')
print('train shape: ', train.shape, '\ntest shape: ', test.shape)

data = pd.concat([train, test])
```

```python
import Preprocess

data_1 = Preprocess.Preprocess(data)
```

```python
data_2 = Preprocess.UserItemShopContext(data_1)
```

```python
output_file = ['round2_uisc_train_7.txt', 'round2_uisc_test_a.txt']

train_uisc = data_2[data_2.is_trade.notnull()]
test_uisc = data_2[data_2.is_trade.isnull()]

train_uisc.to_csv(wd+output_file[0], index=False, sep=' ')
test_uisc.to_csv(wd+output_file[1], index=False, sep=' ')
```
