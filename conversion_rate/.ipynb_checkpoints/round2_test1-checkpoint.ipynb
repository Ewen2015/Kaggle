```python
wd = ['/Users/ewenwang/Documents/practice_data/conversion_rate/', '/Users/ewenwang/Documents/GitHub/Kaggle/conversion_rate/round2/']
file = ['round2_sample_train_uisc.txt', 'round2_test_uisc.txt']

train = pd.read_csv(wd[0]+file[0], sep=' ')
test = pd.read_csv(wd[0]+file[1], sep=' ')
print('train shape: ', train.shape, '/ntest shape: ', test.shape)
```

```python
import lightgbm as lgb

Filter = (train.day==7)
training= train[~Filter]
valid= train[Filter]

target = 'is_trade'
drop_list = ['is_trade', 'item_category_list', 'item_property_list', 'predict_category_property', 'realtime', 'context_timestamp']
features = [x for x in train.columns if x not in drop_list]

X = training[features]
y = training[target].values
X_tes = valid[features]
y_tes = valid[target].values

print('Training LGBM model...')
lgb0 = lgb.LGBMClassifier(
    objective='binary',
    metric='binary_logloss',
    num_leaves=16,
    depth=4,
    learning_rate=0.01,
    seed=2018,
    colsample_bytree=0.6,
    subsample=0.9,
    n_estimators=20000,
    silent = True)

lgb_model = lgb0.fit(X, y, eval_set=[(X_tes, y_tes)], early_stopping_rounds=200)
```

```python
%matplotlib inline
lgb.plot_importance(lgb_model, figsize=(16, 80))
```
